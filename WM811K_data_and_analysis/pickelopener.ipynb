```python
import numpy as np

data = np.load("wm811k_wafer_maps.npz", allow_pickle=True)
wafer_maps = data["wafer_maps"]
```

```python
import matplotlib.pyplot as plt

wafernum = 520476
wafername = str(wafernum)

wafer = wafer_maps[wafernum]   #pick any index you want

plt.figure(figsize=(4, 4))
plt.imshow(wafer, cmap="gray")
plt.colorbar(label="Die value")
plt.title("Wafer: " + wafername)
plt.axis("off")
plt.show()

```
